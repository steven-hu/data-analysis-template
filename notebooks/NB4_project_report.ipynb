**Our Data**

We pulled flight data from 1987 to 2008 from http://stat-computing.org/dataexpo/2009/the-data.html.

The files contain data on all domestic US commercial flights from 1987 to 2008. The data from all years combined consist of over 120 million rows and 29 columns. Each row represents a flight. The columns include (but are not limited to) the following information:

* Year
* Month
* Day of Month
* Day of Week
* Scheduled Departure Time
* Actual Departure Time
* Scheduled Arrival Time
* Actual Arrival Time
* Scheduled Flight Time
* Actual Flight Time
* Departure Delay
* Arrival Delay
* Airport of Origin
* Destination Airport
* Flight Distance
* Flight Number

**Gathering the Data**

Due to the large size of the raw data, we could not work with it directly in iPython Notebook. Instead, we turned to SQLite to construct an SQL database. 

Using the database, we were able to look up and pull the data we wanted as we needed it, instead of attempting to load the data into RAM (which would not have worked due to the sheer size).

**Goals**

Our primary goal was to investigate delays, both by airport and over time. The core question we try to address is:

* Which of the major US airports is the worst, in terms of delayed flights?

There are a couple different metrics that we looked at, including:

* Average delay duration, all flights
* Average delay duration, delayed flights only
* Proportion of flights delayed, 1 minute or more
* Proportion of flights delayed, over _x_ minutes
* Delays over time (1987-2008)

These all are of interest to prospective travelers, but it isn't so easy to determine a metric for "worst" airport. For example, consider airport A that has 80% delayed flights, but an average delay of 2 minutes, versus airport B with 20% delayed flights, but an average delay of 10 minutes. How should we evaluate these airports against each other?

For now, we will simply look at the numbers for each metric, but we will return to the question of evaluation later.

**Data Cleaning**

We pulled the relevant flights for each airport out of the database and saved them as CSV (with python scripts). However, despite filtering out unnecessary columns, we still needed to make some changes based on the delays.

Cancelled flights are removed, flights outside (-60, 500) for delay are removed (removing outliers, in-depth discussion in NB2), and all negative delays (early flights) are set to 0. This is because in evaluating delayed flights, an early flight is essentially the same as an on-time flight. It doesn't make sense to say that an airport that has 3 flights, with delays -5, 0, and 5, has an average delay of 0 minutes.
