**Objective**

In this data analysis project, we investigated flights originating from the 20 busiest United States airports by passenger boarding. Our goal was to determine which of the airports was the "worst" in terms of flight delays. Passengers typically expect their flights to depart and arrive on-time, but time and time again flights are delayed or cancelled. In fact, in the time span between 1987 and 2008, 25 million flights, over 20%, were delayed by at least 15 minutes. That's a lot of frustrated flyers!

There are several different ways to quantify the word "worst." It's difficult to determine a singular metric for determining the "worst" airport, but the metrics we used include the following:

* Average delay duration, all flights
* Average delay duration, delayed flights only
* Proportion of flights delayed, 1 minute or more
* Proportion of flights delayed, over _x_ minutes
* Delays over time (1987-2008)

These metrics all are of interest to prospective travelers, but each individual traveler may find certain metrics more meaningful than others.  For example, consider airport A that has 80% delayed flights, but an average delay of 5 minutes, versus airport B with 10% delayed flights, but an average delay of 50 minutes. Depending on personal preferences, it it difficult to evaluate and make general conclusions when comparing the two airports directly against each other.

For now, we will simply look at the numbers for each metric, but we will return to the question of evaluation later.

**Data Gathering**

* Scheduled Arrival Time
* Actual Arrival Time

* Departure Delay
* Arrival Delay
* Airport of Origin
* Destination Airport
* Flight Distance
* Flight Number

The large size of the raw data made it impossible to work with directly in the iPython Notebook. To set the groundwork for the data analysis, we constructed an SQL database using SQLite.

Using the database, we were able to look up and pull the data we wanted as we needed it, instead of attempting to load the data into RAM (which would not have worked due to the sheer size). The gathering process is documented in more detail in NB1.

**Data Cleaning**

We pulled the relevant flights for each airport out of the database and saved them as CSV (with Python scripts). After filtering out unnecessary columns, we made additional modifications based on the delays.

Cancelled flights are removed, flights outside (-60, 500) for delay are removed to remove outliers (see NB2 for an in-depth discussion), and all negative delays (early flights) are set to 0. This is because in evaluating delayed flights, an early flight is essentially the same as an on-time flight. It doesn't make sense to say that an airport that has 3 flights, with delays -5, 0, and 5, has an average delay of 0 minutes.

**Conclusions**

In our data analysis, we first looked at the average delay times at each airport over the 20-year time span. When we looked at average delay times, including flights that were not delayed (set at 0), we found that the airports with the longest average delays were Chicago O'Hare, Newark Liberty, Philadelphia, and Atlanta, and that those with the shortest average delays were Salt Lake City, George Bush Intercontinental (in Houston), Minneapolis-Saint Paul, and Los Angeles.

When we removed the flights that were not delayed and recalculated the average delay times of only delayed flights, the airports with the longest average delays were Newark, Chicago, Boston Logan, and Detroit, and the airports with the shortest average delays were Pitrsburgh, Salt Lake City, Charlotte, and Los Angeles. Notice that there is some overlap, namely Newark and Chicago for the worst airports and Salt Lake City and Los Angeles for the best airports. This is an indication that the airports' quality based on each of these metrics is not independent: being bad at one is a somewhat good indicator of being bad at another. In fact, as noted above, there is a fairly strong correlation between the two.

Next, we looked at the proportion of flights delayed of all departing flights from each airport over the 20-year time span. We found that the airports with the highest proportion of delayed flights were Atlanta, Pittsburgh, Charlotte, and Chicago, and that airports with the lowest proportion of delayed flights were Minneapolis-Saint Paul, Lambert-St. Louis, La Guardia (New York), and George Bush (Houston).

It was interesting to note that some airports had high average delay times and low proportion of flight delays, and vice versa. For example, Pittsburgh International Airport has the lowest average delay time (with non-delayed times removed from the calculation), but it has the second-highest proportion of flights delayed at more than 50%. Other airports were consistent in both metrics though. Chicago O'Hare International Airport ranks higly in delay times and proportion of flights delayed, and Los Angeles International Airport has both low average delay times and proportion of flights delayed.

The time-series graphs reveal that, between 1987 and 2008, the proportion of flights delays have been generally increasing across the major U.S. airports, and that average flight delay times have been generally decreasing.

It's difficult for us to declare one singular airport as the "worst" among the busiest airports in America, becuase each metric reveals just one of the many facets that determine a timely and consistent airport. But our data anaylsis corroborates our conclusion that Chicago O'Hare and Newark airports perform the worst in terms of average delay times, and Alanta and Pittsburgh airports in terms of proportion of flights delayed. However, Chicago O'Hare (ORD) not only had high average delay times with and without on-time flights removed, it also had a fairly high proportion of flights delayed. 

Depending on how much credence you give to each factor, however, your mileage may vary. For example, the goal of a traveler may be to get to a destination by a certain time. In this case, given that a savvy traveler will budget extra time for contingencies, a high likelihood of a small delay may be tolerable, but the possibility of a very long delay is inexcusable. As a result, this traveler may put the most emphasis on proportion of very long delays when naming the worst airport. Or, an efficiency expert may care only about the overall average delay. Perhaps his or her interests lie in minimizing delay across the board, and thus the raw average statistic is the most important in determining the worst airport.

Simply put, there is no objective measure. But based on our investigation, we'll tentatively name ORD as the worst airport due to being bad by all metrics, with EWR as a very close second simply due to how much higher its delays were compared to all of the other airports.
