Our primary goal was to investigate delays, both by airport and over time. The core question we try to address is:

* Which of the major US airports is the worst, in terms of delayed flights?

Note that since our data encompasses an entire population (domestic commercial flights in the US from 1987-2008), many avenues of statistical inference become unnecessary. There's nothing probabalistic that comes into play when we have a comprehensive view of the population.

We will look at perhaps the simplest measure of "worst" first, the mean delay time at each airport.

## Airline Comparisons




*First, we compared each airport's average delay time per flight:*

```python
%%R
avgs=tapply(data$delay,data$origin,mean)
print(avgs)
```

```python
%%R
hist(avgs,xlab="Average Delay Time (min)",main="Histogram of Average Delay Times")
```

```python
%%R
plot(unique(data$origin),avgs,main="Average Delay Times by Airport",xlab="Airport",ylab="Average Delay Time (min)",xaxt='n')
text(unique(data$origin),avgs,unique(data$origin),cex=0.8, pos=4,col="blue")
abline(h=mean(avgs),lty=2)
text(18, 10.15, "Average delay time", col = "red") 
```

It seems that there's a decent spread between the lowest and highest averages, from about 8 for SLC to about 13 for ORD. However, examining the means is quite simplistic by itself. The inclusion of 0 delay flights (undelayed) in the means lacks some resolution in evaluating the delays per airport. The raw means fail to distinguish between airports with a large proportion of slightly delayed flights and a small proportion of heavily delayed flights. 

In order to look more closely at the data, we'll examine two additional factors:

* Average delay, only including flights that have been delayed (at least 1 minute)
* Proportion of flights delayed

Below, the proportion of flights delayed by airport is plotted.

```python
%%R
carriers=unique(data$origin)
prop={}
for(i in 1:20){
    airport=data$delay[data$origin == carriers[i]]
    total=length(airport)
    del=length(airport[airport != 0])
    prop=c(prop,del/total)
    }
table=data.frame(carriers=carriers, proportiondelayed=prop)
print(table)
```

```python
%%R
plot(table$carriers,table$prop,main="Proportion of Flights Delayed by Airport",xlab="Airport",ylab="Proportion delayed",xaxt='n')
text(table$carriers,table$prop,unique(data$origin),cex=0.8, pos=4,col="blue")
abline(h=mean(table$prop),lty=2)
text(5, 0.449, "Average proportion delayed", col = "red") 
```

*Let's now plot Average Delay Times vs. Proportion of Flights Delayed:*

```python
%%R
cat("Correlation:",cor(table$prop,avgs))
plot(table$prop,avgs,xlab="Proportion of Flights Delayed",ylab="Average Delay Time (min)",xlim=c(0.34,0.56),main="Average Delay Times vs. Proportion of Flights Delayed")
text(table$prop, avgs, carriers, cex=0.6, pos=4,col="blue")
res=lm(avgs~table$prop)
abline(res,lty=2)
```

We see a rough correlation: Higher Proportion of Flights Delayed generally suggests Higher Average Delay Time, although we see exceptions (e.g. SLC and ORD).






*Next, we want to analyze the average delay time of each flight **that was actually delayed**:*

```python
%%R
d=data
d[d==0]=NA
avgs2=tapply(d$delay,d$origin,mean,na.rm=TRUE)
print(avgs2)
```

```python
%%R
plot(unique(data$origin),avgs2,main="Average Delay Times by Airport among Flights Delayed",xlab="Airport",ylab="Average Delay Time (min)",xaxt='n')
abline(h=mean(avgs2),lty=2)
text(unique(data$origin),avgs2,unique(data$origin),cex=0.8, pos=4,col="blue")
text(5, 22.65, "Average delay time", col = "red") 
```

*Let's now compare with this the Average Delay Time **among all flights** for an airport:*

```python
%%R
cat("correlation:",cor(avgs,avgs2))
plot(avgs,avgs2,xlab="Average Delay Time of All Flights (min)",ylab="Average Delay Time (delay > 0 min)",main="Average Delay Time",xlim=c(7.5,13.5))
text(avgs, avgs2, carriers, cex=0.6, pos=4,col="blue")
res=lm(avgs2~avgs)
abline(res,lty=2)
```

We see a rough correlation: Higher Average Delay Time of All Flights generally suggests Higher Average Delay Time for delayed flights, although we see exceptions (e.g. PIT and EWR).

*Let's now compare Average Delay time for delayed flights with proportion of flights delayed:*

```python
%%R
cat('correlation:',cor(table$prop,avgs2))
plot(table$prop,avgs2,xlab="Proportion of Flights Delayed",ylab="Average Delay Time (delay > 0 min)",main="Average Delay Time of Delayed Flights vs Proportion Delayed",xlim=c(0.35,0.55))
text(table$prop, avgs2, carriers, cex=0.6, pos=4,col='blue')
res=lm(avgs2~table$prop)
abline(res,lty=2)
```

Very interesting! We see the greater proportion of flights delayed actually suggests lower average delay time for flights delayed! This is in contrast with the positive correlation we saw earlier of average delay time **for all flights** vs. proportion of flights delayed.

*Finally, let's plot everything side by side for comparsion:*

```python
%%R
par(mfrow=c(1,3))
plot(unique(data$origin),avgs,main="Delay Time: All Flights",xlab="Airport",ylab="Average Delay Time (min)",xaxt='n')
text(unique(data$origin),avgs,unique(data$origin),cex=0.8, pos=3,col="blue")
plot(unique(data$origin),avgs2,main="Delay Time: Flights Delayed",xlab="Airport",ylab="Average Delay Time (min)",xaxt='n')
text(unique(data$origin),avgs2,unique(data$origin),cex=0.8, pos=3,col="blue")
plot(table$carriers,table$prop,main="Proportion Delayed",xlab="Airport",ylab="Proportion delayed",xaxt='n')
text(table$carriers,table$prop,unique(data$origin),cex=0.8, pos=3,col="blue")
```

Next, let's examine the spread of delays at each airport. What we're interested in is how often delays of various severities occur. So, find proportions for each airport of flights delayed at least _x_ minutes, as _x_ goes from 1 to 60.

*Note that for the following section the code wasn't originally run in this notebook, so to avoid the long runtime we've included the code but imported the output.*

```python
%%R
airport = unique(data$origin)
```

```python
%%R
# NOTE: THIS CODE TAKES A LONGGGG TIME TO RUN. TRY TO AVOID USING IT AGAIN. THE LIST IT PULLS, a, IS IN data/cleaned as "output"
a = replicate(20, c())
names(a) = airport
for (n in 1:60) {
    for (i in airport) {  
        a[[i]] = c(a[[i]], sum(data$delay[data$origin == i] >= n)/length(data$delay[data$origin == i]))
    }
}

props = c()
for (n in 1:60) {
    props = c(props, sum(airports$V4 >= n)/length(airports$V4))
}
```

The code above pulls a list of vectors, each vector representing an airport and consisting of the proportion of flights from that airport delayed more than x minutes, x ranging from 1 to 60. It also pulls overall proportions of flights delayed over x minutes, x ranging from 1 to 60.

First, let's plot the overall proportions.

## Time Series Analysis

*We want to analyze the trend of average delay time among flights over the time period (1987-2008). We used the SMA() function in the 
"TTR" R package to fit a smooth series data using a simple moving average. We are trying to find the trend component of this time series accounting for random fluctations.*

```python
%%R
t1987=airports$V4[airports$V2 == 1987]
t1988=airports$V4[airports$V2 == 1988]
t1989=airports$V4[airports$V2 == 1989]
t1990=airports$V4[airports$V2 == 1990]
t1991=airports$V4[airports$V2 == 1991]
t1992=airports$V4[airports$V2 == 1992]
t1993=airports$V4[airports$V2 == 1993]
t1994=airports$V4[airports$V2 == 1994]
t1995=airports$V4[airports$V2 == 1995]
t1996=airports$V4[airports$V2 == 1996]
t1997=airports$V4[airports$V2 == 1997]
t1998=airports$V4[airports$V2 == 1998]
t1999=airports$V4[airports$V2 == 1999]
t2000=airports$V4[airports$V2 == 2000]
t2001=airports$V4[airports$V2 == 2001]
t2002=airports$V4[airports$V2 == 2002]
t2003=airports$V4[airports$V2 == 2003]
t2004=airports$V4[airports$V2 == 2004]
t2005=airports$V4[airports$V2 == 2005]
t2006=airports$V4[airports$V2 == 2006]
t2007=airports$V4[airports$V2 == 2007]
t2008=airports$V4[airports$V2 == 2008]
combined=c(t1987,t1988,t1989,t1990,t1991,t1992,t1993,t1994,t1995,t1996,t1997,t1998,t1999,t2000,t2001,t2002,t2003,t2004,t2005,t2006,t2007,t2008)
timeseries <- ts(combined)
library("TTR")
timeseriesSMA3 <- SMA(timeseries,n=100)
plot.ts(timeseriesSMA3)
```

From the graph, it appears that the length of average delay time has clearly increased over all observations (~ 6.5 million!) over time 

*Next, we used the HoltWinters() function in R to find a simple exponential smoothing predicitive model:*

```python
%%R
expsmooth <- HoltWinters(timeseries, beta=FALSE, gamma=FALSE)
expsmooth

```

The key value we are interested in is alpha. It's value of ~.0329 tells that slightly more weight is based on recent observations then less recent observations (This makes sense from the above graph, since there appears to be a clearly increasing trend) 

*Let's finally look at the SSE (Sum of Squared Errors) of the time series:*

```python
%%R
print(expsmooth$SSE)
```

## ** Logistic regression using log-odds success **

*We are using a generalized linear model with log-odds success of proportions flights delayed as function over time. The linear probability model is not a good fit becuase probabilities are bound between 0 and 1.
We used the simplest logit function: log(p/(1-p)), where p is defined as the proportion of flights delayed.*
    
See: http://pages.uoregon.edu/aarong/teaching/G4075_Outline/node16.html

```python
%%R
log = airports$V4[airports$V2 == 1987 & airports$V4>0]
num = airports$V4[airports$V2 == 1987]
odds1987 = length(log)/length(num)
log = airports$V4[airports$V2 == 1988 & airports$V4>0]
num = airports$V4[airports$V2 == 1988]
odds1988 = length(log)/length(num)
log = airports$V4[airports$V2 == 1989 & airports$V4>0]
num = airports$V4[airports$V2 == 1989]
odds1989 = length(log)/length(num)
log = airports$V4[airports$V2 == 1990 & airports$V4>0]
num = airports$V4[airports$V2 == 1990]
odds1990 = length(log)/length(num)
log = airports$V4[airports$V2 == 1991 & airports$V4>0]
num = airports$V4[airports$V2 == 1991]
odds1991 = length(log)/length(num)
log = airports$V4[airports$V2 == 1992 & airports$V4>0]
num = airports$V4[airports$V2 == 1992]
odds1992 = length(log)/length(num)
log = airports$V4[airports$V2 == 1993 & airports$V4>0]
num = airports$V4[airports$V2 == 1993]
odds1993 = length(log)/length(num)
log = airports$V4[airports$V2 == 1994 & airports$V4>0]
num = airports$V4[airports$V2 == 1994]
odds1994 = length(log)/length(num)
log = airports$V4[airports$V2 == 1995 & airports$V4>0]
num = airports$V4[airports$V2 == 1995]
odds1995 = length(log)/length(num)
log = airports$V4[airports$V2 == 1996 & airports$V4>0]
num = airports$V4[airports$V2 == 1996]
odds1996 = length(log)/length(num)
log = airports$V4[airports$V2 == 1997 & airports$V4>0]
num = airports$V4[airports$V2 == 1997]
odds1997 = length(log)/length(num)
log = airports$V4[airports$V2 == 1998 & airports$V4>0]
num = airports$V4[airports$V2 == 1998]
odds1998 = length(log)/length(num)
log = airports$V4[airports$V2 == 1999 & airports$V4>0]
num = airports$V4[airports$V2 == 1999]
odds1999 = length(log)/length(num)
log = airports$V4[airports$V2 == 2000 & airports$V4>0]
num = airports$V4[airports$V2 == 2000]
odds2000 = length(log)/length(num)
log = airports$V4[airports$V2 == 2001 & airports$V4>0]
num = airports$V4[airports$V2 == 2001]
odds2001 = length(log)/length(num)
log = airports$V4[airports$V2 == 2002 & airports$V4>0]
num = airports$V4[airports$V2 == 2002]
odds2002 = length(log)/length(num)
log = airports$V4[airports$V2 == 2003 & airports$V4>0]
num = airports$V4[airports$V2 == 2003]
odds2003 = length(log)/length(num)
log = airports$V4[airports$V2 == 2004 & airports$V4>0]
num = airports$V4[airports$V2 == 2004]
odds2004 = length(log)/length(num)
log = airports$V4[airports$V2 == 2005 & airports$V4>0]
num = airports$V4[airports$V2 == 2005]
odds2005 = length(log)/length(num)
log = airports$V4[airports$V2 == 2006 & airports$V4>0]
num = airports$V4[airports$V2 == 2006]
odds2006 = length(log)/length(num)
log = airports$V4[airports$V2 == 2007 & airports$V4>0]
num = airports$V4[airports$V2 == 2007]
odds2007 = length(log)/length(num)
log = airports$V4[airports$V2 == 2008 & airports$V4>0]
num = airports$V4[airports$V2 == 2008]
odds2008 = length(log)/length(num)
years=c(1987,1988,1989,1990,1991,1992,1993,1994,1995,1996,1997,1998,1999,2000,2001,2002,2003,2004,2005,2006,2007,2008)
logger2 = c(odds1987/(1-odds1987),odds1988/(1-odds1988),odds1989/(1-odds1989),odds1990/(1-odds1990),odds1991/(1-odds1991),odds1992/(1-odds1992),odds1993/(1-odds1993),odds1994/(1-odds1994),odds1995/(1-odds1995),odds1996/(1-odds1996),odds1997/(1-odds1997),odds1998/(1-odds1998),odds1999/(1-odds1999),odds2000/(1-odds2000),odds2001/(1-odds2001),odds2002/(1-odds2002),odds2003/(1-odds2003),odds2004/(1-odds2004),odds2005/(1-odds2005),odds2006/(1-odds2006),odds2007/(1-odds2007),odds2008/(1-odds2008))
logger=log(logger2)
trend=glm(logger~years)
trend
summary(trend)
```

Thus, we can see time is a significant factor on effect of log-odds success on proportions of flights delayed. There is a significant **decrease** (since the slope from the coefficients, years, is estimated to be negative from above, and the p-value is significant (p=.00575<.01)) in proportion of flights delayed over time. Once again, this is very interesting because we previously saw from the graph above the average length of delays has **increased** over time. Overall, there appear to be fewer delays but the delays are getting longer over time.
