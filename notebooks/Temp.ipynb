Import the cleaned data:

```python
%%R
airports = read.csv("../data/cleaned/data.csv", header=F)
```

Our primary goal was to investigate delays, both by airport and over time. The core question we try to address is:

* Which of the major US airports is the worst, in terms of delayed flights?

Note that since our data encompasses an entire population (domestic commercial flights in the US from 1987-2008), many avenues of statistical inference become unnecessary. There's nothing probabalistic that comes into play when we have a comprehensive view of the population.

Let's look at perhaps the simplest measure of "worst" first, the mean delay time at each airport. Below, the means for each airport's delays are plotted.

```python
%%R -h 500 -w 1000
avgs = tapply(airports$V4, airports$V1, mean)
plot(unique(airports$V1),avgs,main="Average Delay Times by Airport",xlab="Airport",ylab="Average delay times (min)")
abline(h=mean(avgs),lty=2)
```

To get an idea of which airports are better and which are worse by this standard, we'll sort them from lowest to highest mean.

```python
%%R -h 500 -w 1000
sortedavgs = sort(avgs)
name = factor(names(sortedavgs), levels=names(sortedavgs), ordered=T)
plot(name,sortedavgs,main="Average Delay Times by Airport, Lowest to Highest",xlab="Airport",ylab="Average delay times (min)")
abline(h=mean(sortedavgs),lty=2)
```

It seems that there's a decent spread between the lowest and highest averages, from about 8 for SLC to about 13 for ORD. However, examining the means is quite simplistic by itself. The inclusion of 0 delay flights (undelayed) in the means lacks some resolution in evaluating the delays per airport. The raw means fail to distinguish between airports with a large proportion of slightly delayed flights and a small proportion of heavily delayed flights. 

In order to look more closely at the data, we'll examine two additional factors:

* Average delay, only including flights that have been delayed (at least 1 minute)
* Proportion of flights delayed

First, the average delay times of delayed flights only, and overlay the overall averages.

```python
%%R -h 500 -w 1000
avgs2 = tapply(airports$V4[airports$V4 > 0], airports$V1[airports$V4 > 0], mean)
plot(unique(airports$V1),avgs2,main="Average Delay Times by Airport",xlab="Airport",ylab="Average delay times (min)")
abline(h=mean(avgs2),lty=2)
```

Same data sorted:

```python
%%R -h 500 -w 1000
sortedavgs2 = sort(avgs2)
name = factor(names(sortedavgs2), levels=names(sortedavgs2), ordered=T)
plot(name,sortedavgs2,main="Average Delay Times by Airport, Lowest to Highest",xlab="Airport",ylab="Average delay times (min)")
abline(h=mean(sortedavgs2),lty=2)
```

```python
%%R
props = c()
for (n in 1:60) {
    props = c(props, sum(airports$V4 >= n)/length(airports$V4))
}
```

```python
%%R
plot(props, type="l", main="Proportion of Delayed Flights",xlab="Minimum Delay Time (minutes)",ylab="Proportion")
```

```python
%%R
airport = unique(airports$V1)
```

```python
%%R
# NOTE: THIS CODE TAKES A LONGGGG TIME TO RUN. TRY TO AVOID USING IT AGAIN. THE LIST IT PULLS, a, IS IN data/cleaned as "output"
a = replicate(20, c())
names(a) = airport
for (n in 1:60) {
    for (i in airport) {  
        a[[i]] = c(a[[i]], sum(airports$V4[airports$V1 == i] >= n)/length(airports$V4[airports$V1 == i]))
    }
}
```

The code above pulls a list of vectors, each vector representing an airport and consisting of the proportion of flights from that airport delayed more than x minutes, x ranging from 1 to 60.

```python
%%R -h 700 -w 700
color = rainbow(20)
plot(a[[1]], type="l", main="Proportion of Delayed Flights",xlab="Minimum Delay Time (minutes)",ylab="Proportion",col=color[1])
for (i in 2:20) {
    points(1:60, a[[i]], type="l", col=color[i-1])
}
```

It's hard to make out individual lines, but we can see that the airports largely follow the same curve, with different proportions. There are a few crossovers, but no large deviations. The major difference in the chart between lines is vertical: thus some have more delays and some have less, but no airports have a markedly different distribution of delays. EXCEPT: at the high end. We'll revisit this.

So, we plot the proportion of delayed flights for each airport.

```python
%%R -h 500 -w 1000
l = c()
for (i in 1:20) {
    l = c(l, a[[i]][1])
}
plot(airport,l,main="Proportion of Flights Delayed by Airport",ylab="Proportion",xlab="Airports")
abline(h=mean(l),lty=2)
```

And the same plot sorted:

```python
%%R -h 500 -w 1000
names(l) = airport
sortedl = sort(l)
name = factor(names(sortedl), levels=names(sortedl), ordered=T)
plot(name,sortedl,main="Proportion of Delayed Flights, Lowest to Highest",xlab="Airport",ylab="Proportion of Delayed Flights")
abline(h=mean(sortedl),lty=2)
```

This, coupled with the average delay plot, reveal a somewhat different picture of the airports.

Now, back to the rainbow plot. While the shapes of the curves are all very similar, a visual examination makes it seem like there is a substantial deviation on the high end. Let's plot a segment of the data on different axes, to get a better view.

```python
%%R -h 700 -w 700
color = rainbow(20)
plot(30:60,a[[1]][30:60], type="l", main="Proportion of Delayed Flights",xlab="Minimum Delay Time (minutes)",ylab="Proportion",ylim=c(0,.15),col=color[1])
for (i in 2:20) {
    points(30:60, a[[i]][30:60], type="l", col=color[i-1])
}
```

Well, the difference that seemed potentially dramatic no longer looks so dramatic. All the same, the difference in proportion of delays over 1 hour long still looks to be about double between the highest and lowest.
